## Bar plot 

```python
categories = ['A','B','C','D']
values = [10,7.5,15,20]
plt.bar(categories, values, color = 'g')
plt.title('bar plot')
plt.xlabel('categories')
plt.ylabel('values')
plt.show()
```

## Subplot

```python
x1 = np.linspace(0.0, 1.5)
y1 = np.sin(2 * np.pi * x1) * np.exp(-x1)
x2 = np.linspace(0.0, 2.0)
y2 = np.cos(2 * np.pi * x2)

fig, (ax1, ax2) = plt.subplots(2, 1)
ax1.set_title("sin function")


ax1.plot(x1, y1, color = 'r')


ax2.plot(x2, y2)
ax2.set_title("cos function")
ax2.set_xlabel('x-axis')
ax2.set_ylabel('y-axis')

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

plt.figure(figsize=(12, 8))
## write your code here
plt.legend([
"Constant - O(1)", "logarthmic - O(log n)", "Linear O(n)", "quadratic - O(n^2)", " log linear - O(n log n)", "polynomial - O(n^3)","Exponential - O(2^n)"
])
n = np.linspace(1,20,100)

O_1 = np.ones_like(n)
O_log_n = np.log(n)
O_n = n
O_n_squared = n**2
O_n_log_n = n * np.log(n)
O_n_cubed = n**3
O_n_exponetial = 2**n

plt.plot(n, O_1, label='O(1)')
plt.plot(n, O_log_n, label='O(log n)')
plt.plot(n, O_n, label='O(n)')
plt.plot(n, O_n_squared, label='O(n^2)')
plt.plot(n, O_n_log_n, label='O(n log n)')
plt.plot(n, O_n_cubed, label='O(n^3)')
plt.plot(n, O_n_exponetial, label='O(2^n)')






# keep this
plt.ylim(1, 10**7)
plt.yscale("log")  
plt.xlabel("Input Size n", fontsize=12)
plt.ylabel("# Computations", fontsize=12)
plt.title("Complexity Growth of Different Time Complexities", fontsize=16)
plt.legend()
plt.grid(True, which="both", linestyle="--", linewidth=0.5)

plt.show()
```
